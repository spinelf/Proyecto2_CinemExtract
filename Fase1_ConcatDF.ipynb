```python
#Importamos las librerias necesarias
import pandas as pd

# Configuraciones
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)  # Establece una opción de Pandas para mostrar todas las columnas de un DataFrame.
```

```python
#creamos las variables necesarias
genere_list = ['Action', 'Comedy','Drama']
```

```python
#Leemos los 3 csv existentes de los tres géneros y os guardamos en tres variables
genere = genere_list[0]
df_action = pd.read_csv(f'https://raw.githubusercontent.com/fernandaMarti/Proyecto-da-promo-H-modulo-2-team1-cine_freaks/main/Fase1_csv/Fase1_2010_to_2024_{genere}.csv')

genere = genere_list[1]
df_comedy = pd.read_csv(f'https://raw.githubusercontent.com/fernandaMarti/Proyecto-da-promo-H-modulo-2-team1-cine_freaks/main/Fase1_csv/Fase1_2010_to_2024_{genere}.csv')

genere = genere_list[2]
df_drama = pd.read_csv(f'https://raw.githubusercontent.com/fernandaMarti/Proyecto-da-promo-H-modulo-2-team1-cine_freaks/main/Fase1_csv/Fase1_2010_to_2024_{genere}.csv')
```

```python
#Juntamos los 3 DF en uno
frames = [df_action,df_comedy,df_drama]
df_fase1 = pd.concat(frames, ignore_index=True)

#nos aseguramos que elnombre de las columnas es el correcto
df_fase1.columns = ['tipo_pelicula', 'titulo_pelicula', 'anno_estreno', 'mes_estreno', 'id_pelicula','genero_pelicula']
```

```python
#Obtenemos la informacion general del DataFrame
df_fase1.info()
```

```python
#Eliminamos los duplicados
df_fase1.drop_duplicates(subset='id_pelicula', inplace=True)

#Reemplazamos los 'NaN' por 'N/A'
df_fase1.fillna('N/A', inplace=True)

#Reemplazamos en 'mes_estreno' los 'por estrenar' por 'N/A'
df_fase1.loc[df_fase1['mes_estreno']=='por estrenar','mes_estreno']='N/A'
```

```python
#mostramos 10 películas random
df_fase1.sample(10)
```

```python
#Volvemos a ver la informacion general del DataFrame
df_fase1.info()
```

```python
# Vemos estadística de la infomacion
df_fase1.describe(include='object').T
```

```python
#Nos aseguramos que no hay duplicados
df_fase1.duplicated().sum()
```

```python
#Guardamos el DF en formato csv
save_path = f'https://raw.githubusercontent.com/fernandaMarti/Proyecto-da-promo-H-modulo-2-team1-cine_freaks/main/Fase1_csv/Fase1_2010_to_2024_all_generes.csv'
df_fase1.to_csv(save_path, index=False)
```
